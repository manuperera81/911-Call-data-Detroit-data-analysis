#### Convert data types and find null string


```python
df_call_data['called_at'] = pd.to_datetime(df_call_data['called_at'])
```

```python
df_call_data['zip_code'] = df_call_data['zip_code'].replace(r'^\s*$', np.nan, regex=True)
```

```python
# find null strings
df_call_data['zip_code'] = df_call_data['zip_code'].replace(r'^\s*$', np.nan, regex=True)
df_call_data['call_description'] = df_call_data['call_description'].replace(r'^\s*$', np.nan, regex=True)
df_call_data['category'] = df_call_data['category'].replace(r'^\s*$', np.nan, regex=True)
df_call_data['zip_code'] = df_call_data['zip_code'].replace('0    ', np.nan)
```

```python
#updated Null columns
df_call_data.isnull().sum()
```

#### Handle missing/null values

```python
df_call_data['result'] = df_call_data['X'].ge(df_call_data['longitude']) | np.isclose(df_call_data['X'], 
                                                                                            df_call_data['longitude'])
```

```python
df_call_data.head()
```

```python
df_call_data.dropna(subset=['longitude'], inplace=True)
```

```python
df_call_data.isnull().sum()
```

```python
df_call_data_new = df_call_data[(df_call_data['X'].isnull()) & (df_call_data['longitude'].notnull())]
# we have null values in X with longitude is not null So we replace them
```

```python
df_call_data_new
# all have not null values in longitude so we gonna drop X and Y column
```

```python
df_call_data = df_call_data.drop(columns=['X', 'Y','result'])
```

```python
# Interplate on dipstch_time
df_call_data['dispatch_time']= df_call_data['dispatch_time'].interpolate()
```

```python
# replace the null with Unknown for neighbourhood_name
df_call_data['neighborhood_name'].fillna("Unknown",inplace=True)
```

```python
df_call_data.isnull().sum()
```

```python
df_call_data['round_lat'] = df_call_data['latitude'].round(2)
df_call_data['round_long'] = df_call_data['longitude'].round(2)
```

```python
df_call_data.head()
```

```python
#details of zipcode values
df_zipcode = df_call_data[['zip_code','round_long','round_lat']]

```

```python
df_zipcode.dropna()
```

```python
df_zipcode_NA =  df_call_data.loc[df_call_data['zip_code'].isna()]
```

```python
df_nullcopy = df_zipcode_NA.copy()
```

```python
df_nullcopy.info()
```

```python
# merge main data with zipcode for find the Null zip code values
df_merge =pd.merge(left = df_nullcopy, right=df_zipcode, right_on=['round_lat', 'round_long'],
                   left_on=['round_lat', 'round_long'], how='left')
```

```python
df_merge.head()
```

```python
df_merge.dropna(subset=['zip_code_y'], inplace=True)
```

```python
df_final_merge=df_merge.drop_duplicates(subset=['incident_id'])
```

```python
df_final_merge['zip_code_x'] = df_final_merge['zip_code_x'].fillna(df_final_merge['zip_code_y'])
```

```python
df_final_merge.head()
```

```python
df_final_merge.isnull().sum()
```

```python
df_final_merge.info()
```

```python
df_mergeConcat = df_final_merge.drop(columns=['zip_code_y','round_lat','round_long'])
```

```python
df_mergeConcat = df_mergeConcat.rename(columns={'zip_code_x':'zip_code'})
```

```python
df_call_data.dropna(subset=['zip_code'], inplace=True)
df_call_data = df_call_data.drop(columns=['round_lat','round_long'])
```

```python
#df_merge.to_csv("./data/Police_Serviced_911_Calls_2024_cleaned.csv") 
pdList = [df_mergeConcat, df_call_data] 
final_df = pd.concat(pdList)
```

```python
final_df.isnull().sum()
```

```python
final_sorted_df =final_df.sort_values('called_at')
```
